```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("AirQualityUCI.csv")
X = dataset.iloc[ :500, 2:11].values
Y = dataset.iloc[ :500, 14].values
print(X[:10])
print(Y[:10])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/4, random_state = 0)
print(X_test)
print(Y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
#regressor1 = LinearRegression()
#regressor1.fit(X1_train, Y1_train)
```

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
X_train_0 = []
for x in X_train:
    X_train_0.append(x[0])
plt.scatter(X_train_0, Y_train, color = 'red')
#线图
plt.plot(X_train_0, regressor.predict(X_train), 'yellow')
plt.show()
```

```python
X_train_1 = []
for x in X_train:
    X_train_1.append(x[1])
plt.scatter(X_train_1, Y_train, color = 'red')
#线图
plt.plot(X_train_1, regressor.predict(X_train), 'yellow')
plt.show()
```
